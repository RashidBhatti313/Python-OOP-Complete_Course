 Write a program to show classes of umbrella and provide the polymorphism mechanism in 
umbrella open(), close(), waterproof().

```python
class Umberalla:
    def open(self):
        pass

    def close(self):
        pass

    def waterproof(self):
        pass

class RegularUmberalla(Umberalla):
    def open(self):
        print("RegularUmberalla is now open. ")

    def close(self):
        print("RegularUmberalla is now closed. ")

    def waterproof(self):
        print("RegularUmberalla is now waterproof.")
    
class CampactUmberalla(Umberalla):
    def open(self):
        print("CompactUmberalla is now open.")

    def close(self):
        print("CampactUmberalla is now closed.")

    def waterproof(self):
        print("CampactUmberalla is now waterproof.")

class GolfUmberalla(Umberalla):
    def open(self):
        print("GolfUmberalla is now open.")

    def close(self):
        print("GolfUmberalla is now closed.")

    def waterproof(self):
        print("GolfUmberalla is now waterproof.")

def demonstrate_polymorphism(umberalla):
    umberalla.open()
    umberalla.close()
    umberalla.waterproof()

regular_Umberalla = RegularUmberalla()
campact_Umberalla = CampactUmberalla()
Golf_Umberalla = GolfUmberalla()

print("Demonstrating Regular Umbrella:")
demonstrate_polymorphism(regular_Umberalla)

print("\nDemonstrating Compact Umbrella:")
demonstrate_polymorphism(campact_Umberalla)

print("\nDemonstrating Golf Umbrella:")
demonstrate_polymorphism(Golf_Umberalla)
```

 Design class for classical, and latest mobiles. Use polymorphism to show the functions of these 
mobiles such as button() or touch(), size(), split_screen(), operating_system(). Etc.

```python
class MobilePhone:
    def button(self):
        pass

    def touch(self):
        pass

    def size(self):
        pass

    def split_screen(self):
        pass

    def operating_system(self):
        pass

class ClassicalMobile(MobilePhone):
    def button(self):
        print("Classical mobile phone has physical buttons.")

    def touch(self):
        print("Classical mobile phone does not have a touch screen.")

    def size(self):
        print("Classical mobile phone has a smaller screen size.")

    def split_screen(self):
        print("Classical mobile phone does not support split-screen mode.")

    def operating_system(self):
        print("Classical mobile phone runs on a basic operating system.")

class LatestMobile(MobilePhone):
    def button(self):
        print("Latest mobile phone may have virtual buttons or no physical buttons.")

    def touch(self):
        print("Latest mobile phone has a touchscreen for user interaction.")

    def size(self):
        print("Latest mobile phone has a larger screen size with high resolution.")

    def split_screen(self):
        print("Latest mobile phone supports split-screen mode for multitasking.")

    def operating_system(self):
        print("Latest mobile phone runs on a modern operating system (e.g., Android, iOS).")

def demonstrate_polymorphism(phone):
    phone.button()
    phone.touch()
    phone.size()
    phone.split_screen()
    phone.operating_system()

classicalmobile = ClassicalMobile()
latestmobile = LatestMobile()

print("demonstrating Classical Mobile Phones. ")
demonstrate_polymorphism(classicalmobile)

print("\n Demonstrating Latest Mobile Phones. ")
demonstrate_polymorphism(latestmobile)
```

 Design class for Butterfly. Such that butterfly has multiple colors, patters, nature of living, origin, 
living_in_pupa, living_as_caterpiller. Etc

```python
class Butterfly:
    def __init__(self, name, colors, patterns, nature_of_living, origin):
        self.name = name
        self.colors = colors
        self.patterns = patterns
        self.nature_of_living = nature_of_living
        self.origin = origin
        self.is_living_in_pupa = False
        self.is_living_as_caterpillar = False

    def enter_pupa_stage(self):
        self.is_living_in_pupa = True
        self.is_living_as_caterpillar = False

    def enter_caterpillar_stage(self):
        self.is_living_as_caterpillar = True
        self.is_living_in_pupa = False

    def describe(self):
        print(f"Name: {self.name}")
        print(f"Colors: {', '.join(self.colors)}")
        print(f"Patterns: {', '.join(self.patterns)}")
        print(f"Nature of Living: {self.nature_of_living}")
        print(f"Origin: {self.origin}")
        if self.is_living_in_pupa:
            print("Currently living in the pupa stage.")
        elif self.is_living_as_caterpillar:
            print("Currently living as a caterpillar.")
        else:
            print("Currently in an unknown stage of development.")

# Create instances of Butterfly
butterfly1 = Butterfly("Monarch Butterfly", ["Orange", "Black", "White"], ["Striped", "Spotted"],
                       "Flies and pollinates flowers", "North America")

butterfly2 = Butterfly("Swallowtail Butterfly", ["Yellow", "Black"], ["Striped", "Spotted"],
                       "Flies and feeds on nectar", "Europe, Asia, Africa, and the Americas")

# Describe the butterflies
print("Butterfly 1:")
butterfly1.describe()

print("\nButterfly 2:")
butterfly2.describe()

# Simulate lifecycle stages
butterfly1.enter_pupa_stage()
butterfly2.enter_caterpillar_stage()

print("\nUpdated Butterfly 1:")
butterfly1.describe()

print("\nUpdated Butterfly 2:")
butterfly2.describe()
```
