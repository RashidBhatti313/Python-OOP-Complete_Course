```python
class Animal:
    def __init__(self, name, age) -> None:
        self.name = name
        self.age = age

    def speak(self):
        pass

    def walk(self):
        pass

    def Hunting(self):
        pass

    def eat(self):
        pass

    def __str__(self) -> str:
        return f" Name = {self.name} \n Age = {self.age}"
    
class Herbivours(Animal):
    def __init__(self, name, age) -> None:
        super().__init__(name, age)
    
    def speak(self):
        return f" They are speak is Gaaa! "
    
    def walk(self):
        return f" They are walk! "
    
    def eat(self):
        return f" They eat Grass! "
    
    def __str__(self) -> str:
        return f" {super().__str__()} \n {self.speak()} \n {self.walk()} \n {self.eat()}"
    
class Carnivours(Animal):
    def __init__(self, name, age) -> None:
        super().__init__(name, age)

    def speak(self):
        return f" They are speak is KHAaaa! "
    
    def walk(self):
        return f" They are walk! "
    
    def eat(self):
        return f" They eat Meat! "
    
    def Hunting(self):
        return f" Carnivours Animal are hunting! "
    
    def __str__(self) -> str:
        return f" {super().__str__()} \n {self.speak()} \n {self.walk()} \n {self.eat()} \n {self.Hunting()}"
    
class Omnivours(Animal):
    def __init__(self, name, age) -> None:
        super().__init__(name, age)

    def speak(self):
        return f" They are speaking! "
    
    def walk(self):
        return f" They are walk! "
    
    def eat(self):
        return f" They eat Grass and Meat! "
    
    def Hunting(self):
        return f" Omnivours human hunting by buck!"
    
    def __str__(self) -> str:
        return f" {super().__str__()} \n {self.speak()} \n {self.walk()} \n {self.eat()} \n {self.Hunting()}"
    

animal = Animal("Elephant", "7-Year ")
print(animal)

herbivours = Herbivours("Cow", 4)
print(herbivours)

carnivours = Carnivours("LOIN", 5)
print(carnivours)

omnivours = Omnivours("Human", 22)
print(omnivours)
    
```
