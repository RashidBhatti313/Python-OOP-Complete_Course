```python
import time

class AirportDashboard:
    aeroplane_avabaility = True
    dashboard = [
        "1.Jihad  09:00 am",
        "2.Airblue 07:00 pm",
        "3.Shaheen Air 12:00 am",
        "4.Korean Air 06:00 pm",
        "5.Saudi Air 05:00 am"
    ]

    def display(self):
        for item in self.dashboard:
            time.sleep(0.5)
            print(item)

class AirportPerson:
    def __init__(self, name="", cnic=""):
        self.name = name
        self.cnic = cnic

    def get_name(self):
        return self.name
    
    def set_name(self, name):
        self.name = name

    def get_cnic(self):
        return self.cnic
    
    def set_cnic(self, cnic):
        self.cnic = cnic

class Bye:
    off = ["Plane is about to take off!", "\n",
        "Close your seatbelt and be ready for the flight to fly\n"]
    
    def take_off(self):
        for segment in Bye.off:
            time.sleep(0.5)
            print(segment, end="")
    
    @staticmethod
    def bye():
        print("\n.........Have a safe journey..........")
        print("================Good Bye===============")

class Decider:
    airlines = {
        1: "1.Jihad  09:00 am",
        2: "2.Airblue 07:00 pm",
        3: "3.Shaheen Air 12:00 am",
        4: "4.Korean Air 06:00 pm"
    }

    @staticmethod
    def decide(decision):
        if decision in Decider.airlines:
            print(Decider.airlines[decision])
        else:
            raise ValueError("Invalid Decision!")
        
    @staticmethod
    def check_urgent(passenger):
        if passenger.urgent.lower() == "yes":
            price = passenger.price * 1.1
            print(f" Price wih urgency: {price}")
            passenger.set_price(price)
        else:
            price(f"Not urgent: {passenger.price}")

    @staticmethod
    def check_passport(passenger):
        if passenger.passport:
            print("Availability")
        else:
            print("Sorry! You can't travel without a passport")
            exit(0)

    @staticmethod
    def check_weight(passenger):
        if passenger.weight_of_luggage > 30:
            extra_charge = (passenger.weight_of_luggage - 30) * 100
            print(f"You have to pay extra for the overweight: {extra_charge}")
            passenger.set_price(passenger.price + extra_charge)

    @staticmethod
    def check_metal(passenger):
        if not passenger.metal:
            print("\n\n.....Security.........")
            print("\t'Metal Found!'")
            print("You have to come with us for an inquiry.")
            exit(0)

class Main:
    @staticmethod
    def main():
        wlcome = Welcome()
        b = AirportDashboard()
        passenger = Passenger()
        bye = Bye()
        wlcome.display()
        wlcome.welcome()
        b.display()
        print("Fare is the same for all: 10000 Price:")
        print("Choose Airline by the number: ")
        while True:
            try:
                decision = int(input())
                Decider.decide(decision)
                break
            except ValueError as e:
                print(str(e))
                print("Enter the airline decision again!")

        passenger.set_name(input("Enter your name: "))
        passenger.set_cnic(input("Enter your CNIC number: "))
        print("Urgent? Yes or No: ?")
        print("If the ticket is urgent, the extra charge is 1000 RS:")
        passenger.set_urgent(input().strip())
        Decider.check_urgent(passenger)
        print("If luggage is greater than 30 Kg, then you have to pay 100 extra per Kg.")
        print("Your luggage weight: ")
        passenger.set_weight_of_luggage(float(input()))
        Decider.check_weight(passenger)
        print("Passport availability: True or false?")
        passenger.set_passport(input().lower() == 'true')
        print("Checking metal....... true or false?")
        passenger.set_metal(input().lower() == 'true')
        Decider.check_passport(passenger)
        Decider.check_metal(passenger)
        print(f"Total price to pay: {passenger.get_price()} $")
        bye.take_off()
        bye.bye()

class Passenger(AirportPerson):
    def __init__(self, name="", cnic="", passport = False, metal=False, urgent="", weight_of_luggage=0.0, price = 10000.0) -> None:
        super().__init__(name, cnic)
        self.passport = passport
        self.metal = metal
        self.urgent= urgent
        self.weight_of_luggage = weight_of_luggage
        self.price = price

    def get_passport(self):
        return self.passport

    def set_passport(self, passport):
        self.passport = passport

    def get_metal(self):
        return self.metal

    def set_metal(self, metal):
        self.metal = metal

    def get_urgent(self):
        return self.urgent

    def set_urgent(self, urgent):
        self.urgent = urgent

    def get_weight_of_luggage(self):
        return self.weight_of_luggage

    def set_weight_of_luggage(self, weight_of_luggage):
        self.weight_of_luggage = weight_of_luggage

    def get_price(self):
        return self.price

    def set_price(self, price):
        self.price = price

class Welcome:
    wel = "|"

    @staticmethod
    def display():
        print("..........Loading..............")
        for _ in range(30):
            time.sleep(0.1)
            print(Welcome.wel, end="")

    @staticmethod
    def welcome():
        print("\n=====================Welcome To Jinnah Airport==============")
        print("===============DashBoard Displaying==============")


if __name__ == '__main__':
    Main.main()
```
