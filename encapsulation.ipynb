```python
class MyClass:
    def __init__(self):
        self.public_attribute = 42         # Public attribute
        self._protected_attribute = 10     # Protected attribute
        self.__private_attribute = "hello" # Private attribute using name mangling
        
    def get_private_attribute(self):
        return self.__private_attribute
    
    def set_private_attribute(self, value):
        self.__private_attribute = value

# Creating an instance of the class
obj = MyClass()

# Accessing public attribute
print(obj.public_attribute)  # Output: 42

# Accessing protected attribute (convention, not enforced by language)
print(obj._protected_attribute)  # Output: 10

# Accessing private attribute using name mangling (not recommended)
print(obj._MyClass__private_attribute)  # Output: hello

# Using getter method to access private attribute
print(obj.get_private_attribute())  # Output: hello

# Using setter method to modify private attribute
obj.set_private_attribute("world")
print(obj.get_private_attribute())  # Output: world
```

```python
class Person:
    def __init__(self, name , age) -> None:
        self.__n = name
        self.__a = age

    def getN(self):
        return self.__n
    
    def setN(self, name):
        self.__n = name
    
    def getA(self):
        return self.__a
    
    def setA(self, age):
        self.__n = age

    def __str__(self) -> str:
        return f" name = {self.__n} age = {self.__a}"

class Men(Person):
    def __init__(self, name, age, size) -> None:
        super().__init__(name, age)   
        self.s = size 
        
    def __str__(self) -> str:
        return super().__str__()
    
p1 = Person("Ali", 77)
print(p1)
p1.setN("USMAN")
print(p1)
m1 = Men("JAVED", 43, 6)
print(m1)

```

```python
class Bankaccount:
    def __init__(self, account_number, balance) -> None:
        self.acc_num = account_number
        self.__balance = balance

    def deposite(self, amount):
        self.__balance += amount

    def Withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Transaction comleted")
        else:
            print("Insufficient funds!")

    def getbalance(self):
        return self.__balance
    
    def __str__(self) -> str:
        return f" Account_Number = {self.acc_num} \n Balance = {self.__balance}"
    
account = Bankaccount(545454545, 78780000000)
print(account)
print(account.Withdraw(4545454))
print(account.getbalance())

```

```python
class Product:
    def __init__(self, name, price) -> None:
        self.name = name
        self.price = price

    def __str__(self) -> str:
        return f" Name = {self.name} \n Price = {self.price}"
    
class Computer(Product):
    def __init__(self, name, price, specification) -> None:
        super().__init__(name, price)
        self.specification = specification

    def __str__(self) -> str:
        return f" {super().__str__()} \n Specification = {self.specification}"
    
class Laptop(Product):
    def __init__(self, name, price, specification) -> None:
        super().__init__(name, price)
        self.specification = specification

    def __str__(self) -> str:
        return f" {super().__str__()} \n Specification = {self.specification}"

class Accessary(Product):
    def __init__(self, name, price, category) -> None:
        super().__init__(name, price)
        self.category = category

    def __str__(self) -> str:
        return f"{super().__str__()} \n category = {self.category}"
    
class ShoppinigCart:
    def __init__(self) -> None:
        self.items = []

    def add_item(self, product, quantity=1):
        self.items.append((product, quantity))

    def calculate_total(self):
        total = 0
        for product, quantity in self.items:
            total += product.price * quantity
        return total
    
    def __str__(self):
        cart_contents = ""
        for product, quantity in self.items:
            cart_contents += f"{product.name} (Qty: {quantity}) - ${product.price * quantity}\n"
        total_amount = self.calculate_total()
        cart_contents += f"Total Amount: ${total_amount}"
        return cart_contents 
    
computer_1 = Computer("Desktop Computer", 800, "High-performance specs")
laptop_1 = Laptop("Gaming Laptop", 1200, "Powerful GPU")
accessory_1 = Accessary("Keyboard", 50, "Peripherals")
accessory_2 = Accessary("Mouse", 30, "Peripherals")
print(computer_1)
print(laptop_1)
print(accessory_1)
print(accessory_2)


cart = ShoppinigCart()
cart.add_item(computer_1)
cart.add_item(laptop_1)
cart.add_item(accessory_1, 2)
cart.add_item(accessory_2)
print(cart)
```
